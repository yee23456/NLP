#利用jieba進行TFIDF關鍵詞分析後再繪製文字雲

```
import jieba.analyse

#輸出文件前20個關鍵詞，默認表20個
keywords=jieba.analyse.extract_tags(doc1,topK=20,withWeight=True)
dic2=dict(keywords)
dic2
```
