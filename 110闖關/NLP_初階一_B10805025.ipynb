```python
import jieba
import jieba.analyse as analyse
  
#製圖模組
import matplotlib.pyplot as plt
from PIL import Image
```

```python
file=open('中國信託新聞.txt','r')
document1=file.read()
jieba.load_userdict('split.txt')#分詞
seg_list = jieba.lcut(document1, cut_all=False)
file3=open(u'stopword.txt','r',encoding='UTF-8')
stops=file3.readlines()
stopslist=[]
for ww in stops:
  stopslist.append(ww.strip())
stopslist.append(" ")#再加入兩個停用詞
stopslist.append("\n")
stopslist.append("「")
stopslist.append('」')
resultlist=[]
for word in seg_list:
  if word not in stopslist:
    resultlist.append(word)
print(resultlist)
file4=open(r'test4.txt','w', encoding='UTF-8') 
list1=" ".join(resultlist)
file4.write(list1)
file4.close()
```

```python
file1=open('test4.txt','r')
doc1=file1.read()
```

```python
import numpy as np
from PIL import Image 

sp_mask1 = np.array(Image.open('1234.jpg'))
```

```python
from wordcloud import WordCloud 
wc = WordCloud(font_path="WORDTYPE.ttf", repeat=True,background_color="white",
               max_words=2000, mask=sp_mask1,colormap="viridis")

wc.generate(doc1)
```

```python
plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc, interpolation='gaussian')
plt.axis("off")
plt.show()
```
