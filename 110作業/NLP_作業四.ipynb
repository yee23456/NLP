```python
import jieba
```

```python
file=open('doc5.txt','r')
document1=file.read()
jieba.load_userdict('splin1.txt')#分詞
seg_list = jieba.lcut(document1, cut_all=False)
file3=open(u'stopwords0.txt','r',encoding='UTF-8')
stops=file3.readlines()
stopslist=[]
for ww in stops:
  stopslist.append(ww.strip())
stopslist.append(" ")#再加入兩個停用詞
stopslist.append("\n")
stopslist.append("「")
stopslist.append('」')
resultlist=[]
for word in seg_list:
  if word not in stopslist:
    resultlist.append(word)
file4=open(r'test4.txt','w', encoding='UTF-8') 
list1=" ".join(resultlist)
file4.write(list1)
file4.close()
```

```python
import jieba.analyse
file3=open('test4.txt','r')
doc2=file3.read()
#輸出文件前5個關鍵詞，默認表20個
keywords=jieba.analyse.extract_tags(doc2,topK=5,withWeight=True)
for key, weight in keywords:
  print('%s %f'%( key, weight))
```

```python
import numpy as np
from PIL import Image 
import jieba.analyse as analyse
import matplotlib.pyplot as plt
from PIL import Image
sp_mask1 = np.array(Image.open('swallow.jpg'))
from wordcloud import WordCloud 
file1=open('test4.txt','r')
doc1=file1.read()
doc1
```

```python
wc = WordCloud(font_path="/content/drive/MyDrive/大學/NLP/WORDTYPE.ttf", repeat=True,background_color="white",
               max_words=2000, mask=sp_mask1,colormap="viridis")

wc.generate(doc1)
plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc, interpolation='gaussian')
plt.axis("off")
plt.show()
```
