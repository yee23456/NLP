#試做
擷取式摘要 重新評分

```
file1=open('作業文章.txt','r')
text=file1.readlines()
for i in range(len(text)):
  text[i] = re.sub(r'["-,!?.，。、]',' ',text[i])
# text = re.sub(r'["-,!?.，。、]',' ',text)
text
```

```
#訓練模型
word_text=[]
for sentence in text:
    seg_list = jieba.cut(sentence)
    word_text.append(' '.join(seg_list))
x_train = []
for i, text in enumerate(word_text):
  # print(i,text)
  word_list = text.split(' ')
  l = len(word_list)
  # print(l)
  word_list[l - 1] = word_list[l - 1].strip()
  x_train.append(word_list)
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(x_train)]
model = Doc2Vec(documents,dm=1,dbow_words=0,vector_size=10, window=2, min_count=1)
```

```
from textrank4zh import TextRank4Sentence
```

```
file=open('作業文章.txt','r')
txt=file.read()
txt
```

```
tr4s = TextRank4Sentence()
tr4s.analyze(txt, lower=True, source = 'all_filters')

word = []
weight = []
print( '摘要：' )
for item in tr4s.get_key_sentences(num=10):
    # print(item.index, item.weight, item.sentence)# index是句子在文本中位置，weight是此句子的重要性
    word.append(item.sentence)
    weight.append(item.weight)
```

```
test_word_text=[]
for i in range(len(word)):
    seg_list = jieba.cut(re.sub(r'["-,!?.，。、]',' ',word[i]))
    test_word_text.append(''.join(seg_list))  
```

```
test_word_text
```

```
import numpy as np
vec = []
for i in range(len(test_word_text)-1):
  vector0 = model.infer_vector(test_word_text[i]) 
  vector1 = model.infer_vector(test_word_text[i+1])
  sim = np.dot(vector0, vector1) / np.linalg.norm(vector0) / np.linalg.norm(vector1)
  vec.append(sim)
# print(answers[vec.index(max(vec))])
vec #similarity
```

```
# vec = similarity， weight = score ，a = 隨機變數
a = 0.6
for i in range(len(vec)-1):
  new_score = a*weight[i+1] - (1-a)*vec[i]
  print(new_score)
```
