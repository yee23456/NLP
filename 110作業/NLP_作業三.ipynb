```python
file1=open('test5.txt','r')
doc1=file1.read()
```

#直接做成文字雲


```python
import numpy as np
from PIL import Image 

sp_mask1 = np.array(Image.open('1234.jpeg'))
```

```python
from wordcloud import WordCloud 
wc = WordCloud(font_path="/content/drive/MyDrive/大學/NLP/WORDTYPE.ttf", repeat=True,background_color="white",
               max_words=2000, mask=sp_mask1,colormap="viridis")

wc.generate(doc1)
```

```python
# wc.to_file("S1.png")  #將此文字雲儲存成檔案

plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc, interpolation='gaussian')
plt.axis("off")
plt.show()
```

#TFIDF關鍵詞分析

```python
import jieba.analyse

#輸出文件前20個關鍵詞，默認表20個
keywords=jieba.analyse.extract_tags(doc1,topK=20,withWeight=True)
keywords
dic1=dict(keywords)
dic1
```

```python
wc.generate_from_frequencies(dic1)#輸入欲繪製文字雲之文件(須為辭典格式)

plt.figure(figsize=(8,6), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc, interpolation='gaussian')  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```

#textrank4zh  的  textrank關鍵詞分析

```python
!pip install textrank4zh
```

```python
from textrank4zh import TextRank4Keyword
tr4w = TextRank4Keyword()
tr4w.analyze(text=doc1, lower=True, window=2)
```

```python
new_key = {}
key = tr4w.get_keywords(20, word_min_len=2)
for i in range(len(key)):
  new_key[f'{key[i].word}'] = key[i].weight
new_key
```

```python
wc.generate_from_frequencies(new_key)#輸入欲繪製文字雲之文件(須為+)

plt.figure(figsize=(8,6), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc, interpolation='gaussian')  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```
